## Handle Missing Value

```python
# City Column
mode_city = df['City'].mode()[0]
df['City'].fillna(mode_city, inplace=True)
```

```python
# Sales Column
 mean_price = df['Sales'].mean()
df['Sales'].fillna(mean_price, inplace=True)
```

```python
# Ship_Mode Column
mode_Ship = df['Ship_Mode'].mode()[0]
df['Ship_Mode'].fillna(mode_Ship,inplace=True)
```

# Data Visualization

## Bar Chart

```python
#Sales by Category
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure()
category_sales.plot(kind="bar")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

## Bar Chart

```python
#Sales by Sub-Category (Top 10)
subcat_sales = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure()
subcat_sales.plot(kind="bar")
plt.title("Top 10 Sub-Categories by Sales")
plt.xlabel("Sub-Category")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

## Line Chart

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)

df["Month"] = df["Order Date"].dt.to_period("M")

monthly_sales = df.groupby("Month")["Sales"].sum()

plt.figure()
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

## Histogram

```python
plt.figure()
sns.histplot(df['Sales'], bins=60, kde=True)
plt.title('Sales Distribution')
plt.show()
```

## Pie Chart

```python
# Region-wise Sales
region_sales = df.groupby("Region")["Sales"].sum()

plt.figure()
plt.pie(region_sales, labels=region_sales.index, autopct="%1.1f%%")
plt.title("Sales Distribution by Region")
plt.show()
```
